# Bivariate Analysis

## MSZoning vs SalePrice

```python
visualizer.plot('MSZoning', 'SalePrice')
```

### Insights:
- Zoning categories play a critical role in determining property prices, with Residential Low and Floating Village commanding higher prices and Residential High associated with more affordable housing.
- Residential Low has a significant number of high-value outliers, suggesting variability in property sizes or amenities within this zoning category.
- Despite its low frequency, FV properties exhibit premium pricing, which could warrant its retention as a separate category.

# LotFrontage vs SalePrice

```python
visualizer.plot('LotFrontage', 'SalePrice')
```

### Insights:
- There seems to be a weak positive trend between LotFrontage and SalePrice. Larger LotFrontage values tend to have slightly higher SalePrice, but the relationship is not strong or definitive.
- Most data points are clustered in the range of LotFrontage between 50 to 80, and SalePrice between 100,000 to 300,000

## LotArea vs SalePrice

```python
visualizer.plot('LotArea', 'SalePrice')
```

### Insights:
- The majority of the data points are concentrated in the lower ranges of both LotArea (below 20,000) and SalePrice (below 400,000).
- This suggests that most properties in the dataset have smaller lot areas and lower sale prices, which is expected in a typical housing dataset.
- A few extreme values for LotArea (above 50,000, up to 215,000) are noticeable. These likely correspond to large lots, such as commercial properties, farms, or estates, which might not align with the trends of typical residential properties.
- Some of these outliers also have relatively low sale prices, which could indicate data entry errors, unusual circumstances, or unique property characteristics.
